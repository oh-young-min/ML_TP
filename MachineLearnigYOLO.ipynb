#**YOLO 학습**

```python
from ultralytics import YOLO

# Load a model
model = YOLO('yolo11n.pt')

results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=20,
    imgsz=640,
    batch=8,
    name="aquarium-exp"
)
```

#**data.yaml 구성 검토**

```python
import yaml
import os

data_yaml_path=None
for dataset.location, dirs, files in os.walk(dataset.location):
  if "data.yaml" in files:
    data_yaml_path=os.path.join(dataset.location,"data.yaml")
    break

if data_yaml_path and os.path.exists(data_yaml_path):
  print(f"data.yaml 위치 : {data_yaml_path}\n")
  with open(data_yaml_path,"r") as f:
    data=yaml.safe_load(f)

    print(f"클래스 수 (nc): {data.get('nc')}")
    print(f"클래스 이름 (names): {data.get('names')}")
    print(f"학습 이미지 경로 (train): {data.get('train')}")
    print(f"검증 이미지 경로 (val): {data.get('val')}")
    print(f"테스트 이미지 경로 (test): {data.get('test', '정의되지 않음')}")
else:
  print("data.yaml 파일을 찾을 수 없습니다.")
```
